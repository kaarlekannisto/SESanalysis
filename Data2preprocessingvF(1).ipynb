```python
#Initial
data1 = pd.read_csv('Data2.csv') #ensure .ipynb notebook is in the same folder as the data set
pd.set_option('display.max_columns', 999)
data1.head()
```

```python
#Currently uselss columns out
data1.drop('IMDB', axis=1, inplace=True)
data1.drop('OBS', axis=1, inplace=True)
data1 = data1.drop(data1.columns[13:], axis=1)
data1.head()
```

```python
#name to name length
data1['NAME_LENGTH']= data1["NAME"].apply(len)
data1.drop(['NAME', 'P_YEAR', 'VIEWS'], axis=1, inplace=True)
data1.head()
```

```python
#Premiere
data1["PREMIERE"]=data1['PREMIERE'].str[3:5]
data1["PREMIERE"]=data1['PREMIERE'].astype(int)
data1 = pd.get_dummies(data1, columns=['PREMIERE'])
data1.head()
```

```python
#Gender

#data1 = pd.get_dummies(data1, columns=['D_GENDER'])

data1.drop(['D_GENDER'], axis=1, inplace=True)
#data1.head()
```

```python
#CO_COUNTRY
cleanup_nums = {"CO_COUNTRY": {"FI": 0}}
data1.replace(cleanup_nums, inplace=True)
data1.CO_COUNTRY = data1.CO_COUNTRY.apply(lambda x: 1 if x != 0 else 0)
data1 
```

```python
#Genre
g_data = data1['GENRE'].str.get_dummies(sep=', ')
g_data.head()
#Old tries
#data = data1['GENRE'].apply(lambda x: pd.Series(x.split(',')))
#data
```

```python
#Merging
data1 = pd.concat([data1, g_data], axis=1)
data1.drop('GENRE', axis=1, inplace=True)
data1.head()
```

```python
data1 = data1[data1['LENGTH']>0]

length60 = data1['LENGTH'].between(0,60)
length90 = data1['LENGTH'].between(60,90)
length120 = data1['LENGTH'].between(90,120)
length180 = data1['LENGTH'].between(120,180)
data1['LENGTH60'] = length60
data1['LENGTH90'] = length90
data1['LENGTH120'] = length120
data1['LENGTH180'] = length180
```

```python
data1.drop(['LENGTH', 'Short'], axis=1, inplace=True)
```

```python
data1.head()
```

```python
data1['REVENUE_LOG'] = np.log(data1['REVENUE']) 
#data1['REVENUE_LOG'] = data1['REVENUE'] 
```

```python
profitable = data1['REVENUE'] > data1['BUDGET']
data1['PROFIT'] = profitable
```

```python
data1.to_csv('Data2v2.csv', index = False)
```
